```python

# Provides Member Linkedin Profile
member = {
    "name": "Vardhan",
    "roles": ["Backend Developer", "Java Engineer"],
    "skills": ["Java", "Spring Boot", "AWS", "MySQL"],
    "companies": ["TCS", "Infosys"]
}

# For Job description
job = {
    "title": "Data Analyst",
    "skills_required": ["SQL", "Python", "Data Analysis"]
}

# Match skills dynamically
matched_skills = list(
    set(member["skills"]) &
    set(job["skills_required"])
)

#  To Calculate Matched Strength
match_ratio = len(matched_skills) / len(job["skills_required"])

print("\n REFERRAL NUDGE\n")
print(f"Job Role: {job['title']}")

if matched_skills:
    print(f"Matching Skills: {', '.join(matched_skills)}")
else:
    print("Matching Skills: None ")

print("\n Who you can reach out to?")
print(
    f"People you’ve previously worked with who have relevant experience in "
    f"{job['title']}"
)

print("\n Message:")

if match_ratio >= 0.6:
    print(
        f"Hi! I came across a {job['title']} role and thought of you because of "
        f"your experience in {', '.join(matched_skills)}. "
        f"Would you be open to a quick conversation?"
    )
else:
    print(
        f"Hi! I saw a {job['title']} opening and thought it might align with "
        f"your broader experience. Sharing in case you're interested."
    )
```
